```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
def getData():
    df = pd.read_csv("iris.data", names=['sepal length','sepal width','petal length','petal width','target'])
    return df
    
    
```

```python

def main():
    
    
    #Step1: Fetch Datam one image
    df=getData()
    print(df.head())
    
    
    
    #Step 2: Define Features, here features are our columns i.e.: 'sepal length', 'sepal width', 'petal length', 'petal width
    features = ['sepal length', 'sepal width', 'petal length', 'petal width']
    x = df.loc[:, features].values
    y = df.loc[:,['target']].values
    x=pd.DataFrame(data=x, columns = features)
    
    #Step 3: It is necessary to normalize the data
    #x = StandardScaler().fit_transform(x) #to bring everything on one scale
    #print("NORMALIZED DATA")
    #x=pd.DataFrame(data=x, columns = features)
    #print(x.head())
    
    
    
    #Step4: Applying PCA
    
    #Step4.1: Finding Mean with each feature
    x_mean=x.mean(axis=0)
    print("\n\nMean of features in Data")
    print(x_mean)

    
    
    #Step4.2: Shifting the origin
    x_new =dict()   
    for feature in features:
        x_new[feature]=x[feature]-x_mean[feature]
    x_new=pd.DataFrame(data=x_new, columns = features)
    print("\n\nShifted points from origin")
    print(x_new.head())
    
    #Step4.3: Finding Covariance Matrix
    x_prod=1
    
    # Step 4.3.1: Finding "\n\n (X1-mean(X1)) * (X2-mean(X2)) then so on
    
    cov_data = {'sepal length':[0.000,0.000,0.000,0.000],'sepal width':[0.000,0.000,0.000,0.000] , 'petal length':[0.000,0.000,0.000,0.000], 'petal width':[0.000,0.000,0.000,0.000]} 
    
    x_prod = pd.DataFrame(cov_data, index =['petal length', 'petal width', 'sepal length', 'sepal width'])
    
    
    #print(x_prod)
    
    #print(x_prod['sepal length']['sepal length'])
    
    for feature1 in features:
        for feature2 in features:
            product = x_new[feature1] * x_new[feature2]
            #product = x_new.head()[feature1] * x_new.head()[feature2]
            #print("column of "+ feature1 + feature2 )
            #print(product.head())
            variance=product.sum()/(len(product)-1)
            #print("variance"+ str(variance))
            x_prod[feature1][feature2]= variance
            #print("x_prod"+feature1+feature2+ str(x_prod[feature1][feature2]))
            
            #print("variance ::" + str(product.sum()/(len(product)-1)))

    print("\n\n\n\nCOVARIANCE MATRIX:")
    print(x_prod)
    
    # Step 4.3.1: Finding Eigen values and vectors
    import scipy.linalg as LA
    results=LA.eig(x_prod)
    
    print("Eigen Values: ")
    
    eigenvals=results[0].real
    eigenvals= pd.DataFrame(eigenvals, columns =['Eigen Vals'])
    
    print(eigenvals)
    
    print("\n\n\nEigen Vectors:")
    print(results[1])
    eigenvectors= results[1]
    
    
    noofcomponents=int(input("Enter no of principal components to retain: "))
    
    print("\n\n\nPrincipal Components:")
    
    i=0
    pcnames=list()
    for i in range(noofcomponents):
        pcname="Prin. Comp" +str(i+1)
        pcnames.append(pcname)
    
    
    eigenvectors= pd.DataFrame(eigenvectors[:,:noofcomponents], columns =pcnames)
    print(eigenvectors)
    
    print(np.dot(x,eigenvectors))
    
```

```python
if __name__=="__main__":
    main()
```
